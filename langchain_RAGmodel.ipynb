BY HUGGINGFACE ENDPOINT FUNCTION

```python
from langchain_huggingface import HuggingFaceEndpoint
from google.colab import userdata
import os

# Retrieve the Hugging Face token from userdata
sec_key = userdata.get('HF_TOKEN')

# Set the environment variable 'HF_Token' using the retrieved token
os.environ['HF_TOKEN'] = sec_key
```

```python
repo_id = "mistralai/Mistral-Small-24B-Instruct-2501"
# Explicitly setting the task to text-generation
llm = HuggingFaceEndpoint(repo_id=repo_id, task="text-generation", max_length=128, temperature=0.7, token=sec_key)
# corrected typo: temprature -> temperature, toekn -> token
```

```python
llm.invoke("what is machine learning")
```

BY HUGGINGFACE PIPELINE FUNCTION

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_id = "gpt2"
model=AutoModelForCausalLM.from_pretrained(model_id)
```

```python
Tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipe=pipeline("text-generation", model=model, tokenizer=Tokenizer, max_new_tokens=100)
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke(input("enter your text "))
```
